```python
def timeEachNodePrint (distances, trafficConditions, timeOfDay, Day='Sun'): 
    filename = "weights{0}{1}.txt".format(timeOfDay, Day)
    file = open(filename, "w+")
    length = len(distances)
    ctr = 0 
    for i in range(length):
        edge = distances[i][0]
        dist = distances[i][1]
        minutes = timeOfDay % 100
        index = int (minutes / 15 + ( (timeOfDay - minutes) - 600 ) / 25 * 40 ) + ctr
        thisCond = trafficConditions[index][2]
        Time = dist / 44
            # speed limit in Knoxville ~ 30 mph is 44 ft/s 
        traffic = trafficFunction(thisCond, Time)
        thisTravelTime = '{0}'.format(Time + traffic)
        file.write("%s %s\n" %(edge, thisTravelTime))
        ctr += 1
    file.close()
```
