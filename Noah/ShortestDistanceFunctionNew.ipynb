```python
# changing node-naming system to numbers and edge-naming to letters, and adding a dictionary to have location
# names instead of node names. 
def shortestdistance ():
    
#inputing the variables for the function

    timeOfDay = input('Enter the time of day in military time: ')
    
    #error check the timeOfDay input. It must be an int. 
    while not timeOfDay.isnumeric():
        timeOfDay = input('Invalid input for time of day. Please enter time of day as integer in military time.')
        
    #type casting timeOfDay from string to int to use in the function. 
    timeOfDay = int(timeOfDay)
    
    
    Day = input('Enter the day: ')
    
    # casting Day to uppercase so Day can be entered as lower, upper, or a mixture and it will not change 
    # how the program runs. 
    Day = Day.upper()
    
    # Grabbing only the first three letters of the day. 
    Day = Day[:3]
    
    #error check the Day input. It must be one of the seven days of the week, and it must be atleast 3 letters. 
    while (Day != 'SUN' and Day!='MON' and Day!='TUE' and Day!='WED' and Day!='THU'and Day!='FRI' 
    and Day!='SAT' and Day!= 'SUN'):
        Day = input('Invalid input for day. Please enter a day of the week: ')
        Day = Day.upper()
        Day = Day[:3]
        
    while len(Day) < 3:
        Day = input('Invalid input for day. Please enter atleast 3 letters: ')
        Day = Day.upper()
        Day = Day[:3]

        
    locations = dict([(('MELROSE', 'MELROSE'), '1'), (('MELROSE', 'VOLUNTEER'), '2'), 
                  (('LAKE', 'MELROSE'), '3'), (('CUMBERLAND', 'VOLUNTEER'), '4'), 
                  (('PEYTON', 'VOLUNTEER'), '5'), (('17', 'CUMBERLAND'), '6'), 
                  (('17', 'WHITE'), '7'), (('17', 'CLINCH'), '8'), 
                  (('16', 'CLINCH'), '9'), (('CLINCH', 'JAMES'), '10'), 
                  (('16', 'WHITE'), '11'), (('JAMES', 'WHITE'), '12'), 
                  (('CUMBERLAND', 'JAMES'), '13'), (('JAMES', 'PEYTON'), '14'), 
                  (('NEYLAND', 'NEYLAND'), '15'), (('HODGES', ''), '1') ]) 
    ## add more landmarks here. 
    
    
    #input1=input()
    #input2=input()
    #input3 = [input1, input2]
    #input3.sort()
    #locations.get((input3[0], input3[1]))
    
    start = input('Enter the starting location. \n If you enter a street intersection, ' +
                  'please enter only the first word of the street names. \n If you enter a landmark, ' +
                  'please enter only the first word of the landmark: ')
    
    # standardizes all input to upper case. 
    start = start.upper()
    
    # .split will split the input into a list of multiple strings by word. So if we input Melrose, Melrose, 
    # we will have ['Melrose', 'Melrose']
    startlist = start.split(", ")
    
    # sorts alphabetically, which is what we have in the dictionary. 
    startlist.sort() 
    
    # retrieves the node name from the street intersection or landmark name
    startnode = locations.get(startlist[0], startlist[1])
    
    print(len(startlist))
    
    # error check the starting location input. It must 1 or 2, and we are checking if the landmark/intersection
    # exists in the dictionary
#     while len(startlist)!=1 or len(startlist)!=2: # or not(startnode):
#         start = input('Invalid starting location. Please enter first words of street intersection or landmark: ')
#         start = start.upper()
#         startlist = start.split(", ")
#         startlist.sort()
#         startnode = locations.get(startlist[0], startlist[1])

        
    end = input('Enter the ending location. If you enter a street intersection, ' +
                  'please enter only the first word of the street names. If you enter a landmark, ' +
                  'please enter only the first word of the landmark: ')
    
    # standardizes all input to upper case. 
    end = end.upper()
    
    # .split will split the input into a list of multiple strings by word. So if we input Melrose, Melrose, 
    # we will have ['Melrose', 'Melrose']
    end = end.split(", ")
    
    # sorts alphabetically, which is what we have in the dictionary. 
    end.sort() 
    
    # retrieves the node name from the street intersection or landmark name
    endnode = locations.get(end[0], end[1])
    
    #error check the ending location input. It must 1 or 2 and checking if landmark/intersection is valid.  
    while len(end)!=1 or len(end)!=2 or not(endnode):
        end = input('Invalid ending location. Please enter first words of street intersection or landmark: ')
        end = end.upper()
        end = end.split(", ")
        end.sort()
        endnode = locations.get(end[0], end[1])
        

    fileName = './trafficFiles/{0}{1}.txt'.format(Day, timeOfDay)
    
    accessMode='w+'
    
    import numpy as np
    
    
    def speedLimit(edge): 
        
        # Cumberland Ave and Clinch Ave and 16th street 
        if edge == 'E' or edge == 'S' or edge == 'I' or edge == 'J' or edge == 'M' or edge == 'K': 
            valmph = 35
            
        # 17th street 
        elif edge == 'F' or edge == 'G' or edge == 'H': 
            valmph = 30
            
        # Volunteer Blvd and White ave and Melrose Ave 
        elif edge == 'C' or edge == 'B' or edge == 'T' or edge == 'N' or edge == 'A' or edge == 'D': 
            valmph = 25
            
        # Philip Fulmer Way and James Agee and Peyton Manning Pass
        elif edge == 'P' or edge == 'R' or edge == 'O' or edge == 'L' or edge == 'Q': 
            valmph = 20
            
        # conversion factor for mph to ft/s
        valfts = valmph * 1.466666667
        
        return valfts
    
# This function returns the travel time between 2 nodes, given the ideal travel time. 
# Each traffic condition will affect the travel time between the 2 nodes. 
    def trafficTime(trafficCond, Time, dist):
    
        def green(ideal):
            
            actual = 0.0396*ideal**2 + 0.5501*ideal + 2.0513
            return actual

        def orange(ideal): 
            
            actual = 0.0042*ideal**2 + 1.5443*ideal + 1.8068 
            return actual 

        def red(ideal):
            
            actual = 1.6967*ideal**2 - 1.2255*ideal + 2.1677
            return actual

        def brown(ideal): 
            actual = -0.5563*ideal**2 + 11.104*ideal - 2.9879
            return actual

        # Functions are in minutes
        Time /= 60 
        
        
        if trafficCond == 'G':
            traffic = green(Time)
        elif trafficCond == 'O':
            traffic = orange(Time)
        elif trafficCond == 'R':
            traffic = red(Time)
        elif trafficCond == 'B':
            traffic = brown(Time)
        else: #trafficCond == 'NA'
            traffic = Time
            
        # Convert back to seconds
        traffic *= 60 
        
        return traffic

    

# This function will run Noah's program given the start and end nodes and then store the 
# results in the given filename. 

    def PathTime(start, end, fileName):
        
        import os

        file2 = open('input.txt', "w")
        file2.write(start + " " + end)
        file2.close()

        text = './bin/project {0} < ./input.txt'.format(fileName)
        p = os.popen(text)
        print(p.read())
        p.close()
        
        
# the distances vector has the form (edge, distance in feet). It will be used to determine the PTime for the 
# trafficFunction
    distances = [('1A', 377), ('2A', 377), ('2B', 528), ('4B', 528), ('2C', 1056), ('5C', 1056), ('1D', 367), 
                 ('4E', 528), ('13E', 528), ('3F', 430), ('6F', 430), ('6G', 400), ('7G', 400), ('7H', 302), 
                 ('8H', 302), ('8I', 528), ('9I', 528), ('9J', 528), ('10J', 528), ('9K', 300), ('11K', 300), 
                 ('10L', 318), ('12L', 318), ('11M', 350), ('4M', 350), ('11N', 528), ('12O', 400), ('13O', 400), 
                 ('13P', 2112), ('14P', 2112), ('5Q', 528), ('14Q', 528), ('14R', 528), ('15R', 528), 
                 ('4S', 528), ('6S', 528), ('11T', 528)]
    
# input1 is a vector that lists each edge with its nodes. It will be used to determine the second node when 
# writing the weights file. 
    input1 = [('A', '1', '2'), ('B', '2', '4'), ('C', '2', '5'), ('D', '1', '3'), ('E', '4', '13'), 
              ('F', '3', '6'), ('G', '6', '7'), ('H', '7', '8'), ('I', '8', '9'), ('J', '9', '10'),
              ('K', '9', '11'), ('L', '10', '12'), ('M', '4', '11'), ('N', '11', '12'), ('O', '12', '13'), 
              ('P', '13', '14'), ('Q', '5', '14'), ('R', '14', '15'), ('S', '4', '6'), ('T', '7', '11')]
    
# input2 is a vector that lists each node with its possible edges. It will be used to determine the number 
# of nodes (the length). This will be printed in the txt file.
    input2 = [('1', 'A', 'D'), ('2', 'A', 'B', 'C'), ('3', 'D', 'F'), ('4', 'B', 'E', 'M', 'S'), ('5', 'C', 'Q'),
              ('6', 'F', 'G'), ('7', 'G', 'H'), ('8', 'H', 'I'), ('9', 'I', 'J', 'K'), ('10', 'J', 'L'), 
              ('11', 'K', 'M', 'N', 'T'), ('12', 'L', 'N', 'O'), ('13', 'E', 'O', 'P'), ('14', 'P', 'Q', 'R'), 
              ('15', 'R')]
    
# Reading the file with the form HodgeNeylandXY.txt that has the traffic condition for each edge from time 
# X to time Y. 
    
    
    trafficFile = 'HodgesNeyland616.txt' #NOTE: change as we record more traffic data and change the traffic file
    
    #converting trafficFile into a vector. 
    trafficConditions = np.loadtxt(trafficFile, dtype=str)
    
    # length is for the for loop below. This is how many data points we will enter. 
    length = len(distances)
    
    # creating new file to write the data in
    file = open(fileName, accessMode)
    
    # len(input2) is the number of nodes. 
    file.write("%s\n" %(len(input2)))
    
    # minute can be 0, 15, 30, 45
    minute = timeOfDay % 100
    
    # index2 will be the index of the trafficConditions vector that we are referring to. 
    # 600 is 0, 615 is 40, ...
    index1 = int (160*(timeOfDay-minute-600)/100 + 40*(minute/15))
    
 
    for i in range(length):
        # edgen is the edge name. It will have the form '2B' or '11I'. 
        edgen = distances[i][0]
        
        # Defining node1, the from node, which is '1', '2', .... We start by saying only the first element in the
        # string is the edge, but we keep adding numbers to our node string until we find an element that is a 
        # letter, taking care of the cases when the node is more than 1 digit. 

        j = 0
        while edgen[j].isnumeric():
            edge = edgen[:j+1]
            j+=1
        
        # Defining edge, which will have the form 'A', 'B', ..., which will be the rest of the edgen vector 
        # that was not used for node1. 
        node1 = edgen[j:] 

        # index1 will be used to access the input1 list. input1 is in order of the edges. we need to find the 
        # ASCII code for the edge to get the index, subtract 65 because 'A' - 'Z' starts at 65, 
        # then subtract 1 because 0-indexing
        index2 = int(ord(edge)) - 66
        
        # node2 will either be the 2nd or 3rd element in the list accessed by input1[index1]. We assign node2 
        # to the 2nd element first. The if statement checks if node2 is the same as node1. If they are, then we 
        # assign node2 to the 3rd element. 
        node2 = input1[index2][1] #to node
        if node2 == node1: 
            node2 = input1[index2][2]
            
        # dist accesses the specific distance at the edge we are working with. It will be the distance we 
        # use to calculate the Ideal Time for the trafficTime function. 
        dist = distances[i][1]
        
        
        # defining thisCond, which will use index2 to access the traffic condition at this edge and this time. 
        thisCond = trafficConditions[index1][2]
        
        # ITime is the ideal travel time. It is the distance divided by the speed limit of the edge
        speed = speedLimit(edge)
        ITime = dist / speed

        # traffic will be the extra time spent driving due to the traffic conditions. 
        traffic = trafficTime(thisCond, ITime, dist)
        
        # incrementing index2 by 1 because trafficCond goes in the same order as the distances list.
        index2 += 1
            
        # writing the traffic condition on the file.  
        file.write("%s %s %s\n" %(node1, node2, traffic))
        
        ## need to change function when we have more than one day's data. ## 
    
    file.close()
    
    # reads the text file written in the for loop, sends into Noah's program. Noah's program will print
    # the fastest route. 
    PathTime(start, end, fileName)
```
